1. Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

```python
from scipy import stats

# Hypotheses
# Null hypothesis (H0): There is no significant difference in MEDV based on CHAS
# Alternative hypothesis (H1): There is a significant difference in MEDV based on CHAS

# Perform the t-test for independent samples
t_stat, p_value = stats.ttest_ind(
    boston_df[boston_df['CHAS'] == 1]['MEDV'],
    boston_df[boston_df['CHAS'] == 0]['MEDV']
)

# Significance level
alpha = 0.05

# Determine whether to reject the null hypothesis
if p_value < alpha:
    conclusion = "Reject the null hypothesis. There is a significant difference in MEDV based on CHAS."
else:
    conclusion = "Fail to reject the null hypothesis. There is no significant difference in MEDV based on CHAS."

# Output the test statistic, p-value, and conclusion
print(f'T-Statistic: {t_stat}')
print(f'P-Value: {p_value}')
print(conclusion)
```

2. Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming 'boston_df' is the DataFrame that contains your data

# Discretize the 'AGE' variable into three groups
boston_df['AGE_GROUP'] = pd.cut(boston_df['AGE'], bins=[-1, 35, 70, float('inf')],
                                labels=['35 years and younger', '35-70 years', 'older than 70 years'])

# Create a boxplot for the MEDV variable vs the AGE_GROUP variable
plt.figure(figsize=(10, 6))
sns.boxplot(x='AGE_GROUP', y='MEDV', data=boston_df)
plt.title('Boxplot of MEDV vs. Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Median Value of Owner-Occupied Homes in $1000\'s')
plt.show()
```

```python
import pandas as pd
from scipy import stats

# Assuming 'boston_df' is your DataFrame and it contains 'MEDV' and 'AGE' columns

# First, you need to categorize the 'AGE' variable into groups
boston_df['AGE_GROUP'] = pd.cut(boston_df['AGE'], bins=[-1, 35, 70, float('inf')],
                                labels=['35 years and younger', '35-70 years', 'older than 70 years'])

# Then perform ANOVA to test the differences among groups
f_value, p_value = stats.f_oneway(
    boston_df.loc[boston_df['AGE_GROUP'] == '35 years and younger', 'MEDV'],
    boston_df.loc[boston_df['AGE_GROUP'] == '35-70 years', 'MEDV'],
    boston_df.loc[boston_df['AGE_GROUP'] == 'older than 70 years', 'MEDV']
)
print("F-Statistic:", f_value)
print("P-Value:", p_value)
# The f_value is the calculated F-statistic for the ANOVA test, and p_value tells us whether the test results are significant or not.
```

The F-Statistic is significantly high, and the P-Value is extremely low (much less than any common alpha level like 0.01 or 0.05). This suggests that there is a statistically significant difference in the median values of houses (MEDV) among the different age categories of owner-occupied units built prior to 1940 (AGE). Therefore, you would reject the null hypothesis that there are no differences in median values across the different age groups

3. Can we conclude that there is no relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'boston_df' is the DataFrame that contains your data

# Create a scatter plot to examine the relationship between NOX and INDUS
plt.figure(figsize=(10, 6))
sns.scatterplot(x='INDUS', y='NOX', data=boston_df)
plt.title('Scatter Plot of Nitric Oxide Concentrations vs. Proportion of Non-Retail Business Acres')
plt.xlabel('Proportion of Non-Retail Business Acres per Town (INDUS)')
plt.ylabel('Nitric Oxide Concentrations (NOX)')
plt.show()
```

```python
from scipy.stats import pearsonr

# Assuming 'boston_df' is your DataFrame containing the NOX and INDUS columns
corr_coefficient, p_value = pearsonr(boston_df['NOX'], boston_df['INDUS'])

print("Pearson Correlation Coefficient:", corr_coefficient)
print("P-Value:", p_value)
```

The correlation coefficient of 0.76 indicates a strong positive relationship between nitric oxide concentrations and the proportion of non-retail business acres per town. This means that as the proportion of non-retail business acres increases, nitric oxide concentrations also tend to increase.

The extremely small P-Value 7.913361061210442e-98 significantly below any conventional alpha level (e.g., 0.05, 0.01), strongly suggests that the correlation observed is statistically significant.

Therefore, we can conclude that there is a relationship between nitric oxide concentrations and the proportion of non-retail business acres per town, rejecting the hypothesis that there is no relationship.








4. What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Assuming 'boston_df' is your DataFrame containing the data

# Prepare the data
X = boston_df[['DIS']]  # Predictor variable: DIS
y = boston_df['MEDV']   # Response variable: MEDV

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Retrieve the intercept and coefficient(slope) of the model
intercept = model.intercept_
slope = model.coef_[0]

print(f"Intercept: {intercept}")
print(f"Slope: {slope}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'boston_df' is your DataFrame containing the DIS and MEDV columns

# Create a scatter plot with a regression line to show the relationship
plt.figure(figsize=(10, 6))
sns.regplot(x='DIS', y='MEDV', data=boston_df, line_kws={"color": "red"})
plt.title('Impact of Weighted Distance to Boston Employment Centers on Median Home Value')
plt.xlabel('Weighted Distance to Five Boston Employment Centers (DIS)')
plt.ylabel('Median Value of Owner-Occupied Homes (MEDV)')
plt.show()
```

These results indicate that the baseline median value of owner-occupied homes (MEDV), when the weighted distance to the five Boston employment centers (DIS) is 0, is approximately $18,390 (assuming the median value is in thousands of dollars, as is common in this dataset).

The positive slope of 1.09 suggests that for each additional unit increase in the weighted distance to the employment centers, the median value of owner-occupied homes increases by approximately $1,090 (again, assuming the values are in thousands).

Therefore, the answer to the question "What is the impact of an additional weighted distance to the five Boston employment centres on the median value of owner-occupied homes?" is that there is a positive relationship between DIS and MEDV. As the distance to employment centers increases, so does the median value of the homes, according to the data and the constructed linear regression model.
